```python
import numpy as np
import scipy as sp
import matplotlib as mpl
from matplotlib import pyplot as plt
from scipy import linalg
from scipy.special import gamma, factorial, binom, comb,i0,iv,ive

## Quadrature ellipse given mean vector, and squeezing parameter d=np.exp(2*np.abs(z)), thet is argument of z

#Mean vector
m=np.array([-.1,.8])

#Consider covariance matrix defined by 
# gg=(d/2)+(1/(2*d))
# hh=(d/2)-(1/(2*d))
# xx=(1/2)*np.array([[gg-(hh*np.cos(thet)),-np.sin(thet)*hh],[-np.sin(thet)*hh[0],gg+(hh[0]*np.cos(thet))]])

#Can solve for d and thet to proceed to plot. C XXVII p.104. If given pure state covariance
#matrix [[a,b],[b,((b**2)+(1/4))/a]], get thet = np.arcsin(np.abs(b)/((1/2)*gg)) and
# d=(1/(2*a))*
#   ((2*(np.abs(b)**2))+(2*(a**2))+(1/2)+np.sqrt( (((2*(np.abs(b)**2)) + (1/2) +(2*(a**2)))**2)-(4*(a**2)) ))

d=3
## Negative correlation goes with positive thet in our convention
thet=.8

dats=[]
dats.append([m,d,thet])
```

```python
dats[0][0]
```

```python
ell=[]
for k in range(len(dats)):
    mean = dats[k][0]
    width = 1/(2*dats[k][1])
    height = 1/width
    rot = 180*dats[k][2]/np.pi
    ell.append(mpl.patches.Ellipse(xy=mean, width=width, height=height, angle = rot))
    
fig, ax = plt.subplots(figsize=(10,10))

for j in range(len(ell)):
    ell[j].set_alpha(0.5)
    col=np.random.rand(3)
    ell[j].set_color(col)
    ax.add_artist(ell[j])
    
    x_values = [0,m[0]]
    y_values = [0,m[1]]
    plt.plot(x_values, y_values)

ax.set_aspect('equal')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
plt.show()
```

```python
#Now we rotate the state
phi=np.pi/4
oo=np.array([[np.cos(phi),-np.sin(phi)],[np.sin(phi),np.cos(phi)]])

m=m@oo

#Consider covariance matrix defined by 
# gg=(d/2)+(1/(2*d))
# hh=(d/2)+(1/(2*d))
# xx=(1/2)*np.array([[gg-(hh*np.cos(thet)),-np.sin(thet)*hh],[-np.sin(thet)*hh[0],gg+(hh[0]*np.cos(thet))]])

d=3
## Negative correlation goes with positive thet in our convention
thet=.8-(2*phi)

dats=[]
dats.append([m,d,thet])
```

```python
ell=[]
for k in range(len(dats)):
    mean = dats[k][0]
    width = 1/(2*dats[k][1])
    height = 1/width
    rot = 180*dats[k][2]/np.pi
    ell.append(mpl.patches.Ellipse(xy=mean, width=width, height=height, angle = rot))
    
fig, ax = plt.subplots(figsize=(10,10))

for j in range(len(ell)):
    ell[j].set_alpha(0.5)
    col=np.random.rand(3)
    ell[j].set_color(col)
    ax.add_artist(ell[j])
    
    x_values = [0,m[0]]
    y_values = [0,m[1]]
    plt.plot(x_values, y_values)

ax.set_aspect('equal')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
plt.show()
```
